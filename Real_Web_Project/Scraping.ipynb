```python
import requests as req
from bs4 import BeautifulSoup
import pandas as pd
```

```python
html_url = req.get("https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=python&txtLocation=").text

Title_List = []
Skills_List = []
Experience_List = []
Post_Time_List=[]
Info_URL_List =[]


unfimiliar_Skills = input("Enter Unfimiliar Skills: ")

soup = BeautifulSoup(html_url, "lxml")
jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")


for job in jobs:
    
    Post_Time = job.find("span", class_= "sim-posted").span.text.strip()
    Skills = job.find("span", class_= "srp-skills").text.replace(" ", "").strip()
    
    if "few" in Post_Time and unfimiliar_Skills not in Skills and unfimiliar_Skills.capitalize() not in Skills :
        Title= job.find("h3", class_= "joblist-comp-name").text.strip()
        Experience = job.find("ul", class_= "top-jd-dtl clearfix").li.text.replace("card_travel","")
        Info_URL = job.header.h2.a["href"]
        
        Title_List.append(Title)
        Skills_List.append(Skills)
        Experience_List.append(Experience)
        Info_URL_List.append(Info_URL)
        Post_Time_List.append(Post_Time)
        
        print(f"{Title}\n{Skills}\n{Post_Time}\n{Experience}\n{Info_URL}\n\n")
        

        
        
```

```python
data = {"Job Title":Title_List,
        "Skills": Skills_List,
        "Experience": Experience_List,
        "Post_Time":Post_Time_List,
        "Info_URL": Info_URL_List
        }

df = pd.DataFrame(data)
df
```

```python
df.to_excel('file1.xlsx')
```
